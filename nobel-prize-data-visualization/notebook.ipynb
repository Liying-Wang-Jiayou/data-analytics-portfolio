```python
# Loading in required libraries
import pandas as pd
import seaborn as sns
import numpy as np

# use pd.read_csv to load nobel dataset into a DataFrame
nobel = pd.read_csv('data/nobel.csv')

# use value_counts() to find the most common value in a column, and use index[0] to grab only the top value, cuz value_counts() goes by descending
top_gender = nobel['sex'].value_counts().index[0]
top_country = nobel['birth_country'].value_counts().index[0]

# create the us_born winners column
nobel["US_born_winner"] = nobel['birth_country'] == 'United States of America'
#create the decade column
nobel["decade"] = (nobel["year"] // 10 * 10).astype(int)

result = nobel.groupby('decade', as_index=False)['US_born_winner'].mean()
max_ratio = result['US_born_winner'].max()
max_decade_usa = result[result['US_born_winner'] == max_ratio]['decade'].values[0]
print(max_decade_usa)

# creating a new column 'Is_Female' to indicate if the winner is female
# 创建 is_female 列
nobel["is_female"] = nobel["sex"] == "Female"

# 计算每个 decade 女性获奖者的比例
female_result = nobel.groupby('decade', as_index=False)['is_female'].mean()

# 找到女性获奖比例最高的 decade
max_female_decade = female_result.loc[female_result['is_female'].idxmax(), 'decade']

# 计算该 decade 下，各类别的女性获奖比例
female_category_ratio = nobel[nobel['decade'] == max_female_decade].groupby('category')['is_female'].mean()

# 选出女性获奖比例最高的类别
max_female_category = female_category_ratio.idxmax()

# 创建正确的字典
max_female_dict = {max_female_decade: max_female_category}

print(max_female_dict)

nobel_woman = nobel[nobel["is_female"]]
first_woman = nobel_woman[nobel_woman["year"] == nobel_woman["year"].min()]
first_woman_name = first_woman["full_name"].values[0]
first_woman_category = first_woman["category"].values[0]

print(first_woman)
print(first_woman_category)

unique_name = nobel['full_name'].value_counts()
two_or_more = unique_name[unique_name >= 2].index
repeat_list = list(two_or_more)
print(repeat_list)
```
